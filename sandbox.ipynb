```python
import pandas as pd
import json
import numpy as np
```

```python


# 1. Tentukan path ke file Load Ready
file_path = "data/reports/Load_Ready_20260401_1344.csv"

print(f"Membaca file dari {file_path} ...")
# 2. Baca CSV
df_ready = pd.read_csv(file_path)
print(f"Total baris awal: {len(df_ready):,}")

# 3. Ekstrak data JSON ke dalam format kolom
print("Parsing Row_Data_JSON...")
# Menggunakan trik dict comprehension & list comprehension untuk parse JSON
parsed_data = [json.loads(row) for row in df_ready["Row_Data_JSON"]]

# 4. Buat DataFrame baru berdasarkan data yang sudah diekstrak
# Lalu gabungkan dengan 'Dataset_Id' dan 'Judul_Tabel' dari yang aslinya
df_extracted = pd.DataFrame(parsed_data)

# Agar kita tahu baris ini asalnya dari dataset mana
df_final = pd.concat([
    df_ready[["Dataset_Id", "Judul_Tabel"]],
    df_extracted
], axis=1)

print("Proses selesai!")
display(df_final.head())
```

```python
df_final.tail()
```

```python
df_filter = df_final[["Dataset_Id", "Judul_Tabel", "tahun"]].copy()
df_filter['tahun'] = df_filter['tahun'].astype("int")
```

```python
df_filter['tahun'].value_counts()
```

```python
df_tahun = df_filter[(df_filter["tahun"] >= 2019) & (df_filter["tahun"] <= 2026)]
df_tahun.sort_values(by="tahun", ascending=False)
df_tahun["tahun"].value_counts()
```

```python
df_encoded = pd.get_dummies(df_tahun, columns=['tahun'])
df_encoded.head()
```

```python
df_combined = df_encoded.groupby(["Judul_Tabel", "Dataset_Id"]).any().reset_index()
cols_to_fix = df_combined.columns.difference(['Judul_Tabel', 'Dataset_Id'])
df_combined[cols_to_fix] = np.where(df_combined[cols_to_fix], "Tersedia", "Tidak Tersedia")
df_combined.info()
```

```python
df_combined.head()
```

```python
df_combined.to_excel("data/reports/data_ketersediaan_per_tahun.xlsx")
```
